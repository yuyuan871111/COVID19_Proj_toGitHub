## do the same thing as main (read data, quick plot and save in two file): CLASS

```python
class do_the_same_thing:
    def __init__(self, do_moving_avg, avg_window, variants_name, xvgfile, colorlist, thetitle, thexlabel, theylabel, datacolumn=1, ns_factor = 10, do_ps2ns = False):
        self.do_moving_avg = do_moving_avg
        self.avg_window = avg_window
        self.datacolumn = datacolumn

        self.variants_name = variants_name
        self.xvgfile = xvgfile

        self.thetitle = thetitle
        self.thexlabel = thexlabel
        self.theylabel = theylabel
        self.ns_factor = ns_factor
        self.colorlist = colorlist

        # quick read data: 7a91
        x_list, y_list, header_list, x_avg_list, y_avg_list = [],[],[],[],[]
        for variant in self.variants_name:
            wkdir = f'metadata/md_results/variants_7a91/{variant}'
            x, y, header, x_avg, y_avg = read_xvg(wkdir=wkdir, xvgfile=self.xvgfile, do_moving_avg=self.do_moving_avg, datacolumn=self.datacolumn, avg_window = self.avg_window, ps2ns=do_ps2ns)
            x_list.append(x)
            y_list.append(y)
            header_list.append(header)
            x_avg_list.append(x_avg)
            y_avg_list.append(y_avg)

        self.pdb_7a91 = {
            "x_list":x_list,
            "y_list":y_list,
            "header_list":header_list,
            "x_avg_list":x_avg_list,
            "y_avg_list":y_avg_list
        }

        # quick read data: 6m0j
        x_list, y_list, header_list, x_avg_list, y_avg_list = [],[],[],[],[]
        for variant in self.variants_name:
            wkdir = f'metadata/md_results/variants_6m0j/{variant}'
            x, y, header, x_avg, y_avg = read_xvg(wkdir=wkdir, xvgfile=self.xvgfile, do_moving_avg=self.do_moving_avg, datacolumn=self.datacolumn, avg_window = self.avg_window, ps2ns=do_ps2ns)
            x_list.append(x)
            y_list.append(y)
            header_list.append(header)
            x_avg_list.append(x_avg)
            y_avg_list.append(y_avg)

        self.pdb_6m0j = {
            "x_list":x_list,
            "y_list":y_list,
            "header_list":header_list,
            "x_avg_list":x_avg_list,
            "y_avg_list":y_avg_list
        }
        
        # quick read data: ground truth
        self.variants_name_groundtruth = ['WT_6m0j', 'WT_7a91', 'alpha_7mjn', 'beta_7v80', 'gamma_7v84', 'delta_7v8b']
        self.colorlist_groundtruth = ['#adadad', '#7d7d7d', '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
        self.variants_groundtruth = {
            'variants_6m0j':'WT',
            'variants_7a91':'WT',
            'variants_7mjn':'alpha',
            'variants_7v80':'beta',
            'variants_7v84':'gamma',
            'variants_7v8b':'delta'
        }
        x_list, y_list, header_list, x_avg_list, y_avg_list = [],[],[],[],[]
        for variant_folder in self.variants_groundtruth:
            wkdir = f'metadata/md_results/{variant_folder}/{self.variants_groundtruth[variant_folder]}'
            x, y, header, x_avg, y_avg = read_xvg(wkdir=wkdir, xvgfile=self.xvgfile, do_moving_avg=self.do_moving_avg, datacolumn=self.datacolumn, avg_window = self.avg_window, ps2ns=do_ps2ns)
            x_list.append(x)
            y_list.append(y)
            header_list.append(header)
            x_avg_list.append(x_avg)
            y_avg_list.append(y_avg)

        self.pdb_groundtruth = {
            "x_list":x_list,
            "y_list":y_list,
            "header_list":header_list,
            "x_avg_list":x_avg_list,
            "y_avg_list":y_avg_list
        }
        
        # create avg with time
        y_lists = [self.pdb_7a91['y_list'], self.pdb_6m0j['y_list'], self.pdb_groundtruth['y_list']]
        self.dfs = []
        for y_list in y_lists:
            try:
                avg_list = []
                for index, _ in enumerate(variants_name):
                    avg_with_time = []
                    
                    for n in range(self.ns_factor):
                        avg_with_time.append( np.mean( np.array(y_list[index][1000*n:1000*(n+1)+1]) ) )
                    
                    avg_list.append(avg_with_time)

                df = pd.DataFrame(avg_list)
                df.index = variants_name
                the_columns = []
                for n in range(self.ns_factor):
                    the_columns.append(f"{theylabel} ({0+10*n}-{10*n+10}ns)")
                df.columns = the_columns
                self.dfs.append(df)
            except:
                avg_list = []
                for index, _ in enumerate(self.variants_name_groundtruth):
                    avg_with_time = []
                    
                    for n in range(self.ns_factor):
                        avg_with_time.append( np.mean( np.array(y_list[index][1000*n:1000*(n+1)+1]) ) )
                    
                    avg_list.append(avg_with_time)

                df = pd.DataFrame(avg_list)
                df.index = self.variants_name_groundtruth
                the_columns = []
                for n in range(self.ns_factor):
                    the_columns.append(f"{theylabel} ({0+10*n}-{10*n+10}ns)")
                df.columns = the_columns
                self.dfs.append(df)
    

    def quickplot(self):
        fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=[27,6])
        # quick plotting 6m0j
        multiple_variants_plot(title=f"{self.thetitle} (6m0j)", xlab=self.thexlabel, ylab=self.theylabel, variants_name=self.variants_name, 
                                x_list=self.pdb_6m0j["x_list"], y_list=self.pdb_6m0j["y_list"], ax=ax1,
                                x_avg_list=self.pdb_6m0j["x_avg_list"], y_avg_list=self.pdb_6m0j["y_avg_list"], 
                                do_moving_avg=False, alpha=0.15, colorlist=self.colorlist, showlabel=False)
        multiple_variants_plot(title=f"{self.thetitle} (6m0j)", xlab=self.thexlabel, ylab=self.theylabel, variants_name=self.variants_name, 
                                x_list=self.pdb_6m0j["x_list"], y_list=self.pdb_6m0j["y_list"], ax=ax1,
                                x_avg_list=self.pdb_6m0j["x_avg_list"], y_avg_list=self.pdb_6m0j["y_avg_list"], 
                                do_moving_avg=self.do_moving_avg, colorlist=self.colorlist)
        
        # quick plotting 7a91           
        multiple_variants_plot(title=f"{self.thetitle} (7a91)", xlab=self.thexlabel, ylab=self.theylabel, variants_name=self.variants_name, 
                                x_list=self.pdb_7a91["x_list"], y_list=self.pdb_7a91["y_list"], ax=ax2,
                                x_avg_list=self.pdb_7a91["x_avg_list"], y_avg_list=self.pdb_7a91["y_avg_list"], 
                                do_moving_avg=False, alpha=0.15, colorlist=self.colorlist, showlabel=False)
        multiple_variants_plot(title=f"{self.thetitle} (7a91)", xlab=self.thexlabel, ylab=self.theylabel, variants_name=self.variants_name, 
                                x_list=self.pdb_7a91["x_list"], y_list=self.pdb_7a91["y_list"], ax=ax2,
                                x_avg_list=self.pdb_7a91["x_avg_list"], y_avg_list=self.pdb_7a91["y_avg_list"], 
                                do_moving_avg=self.do_moving_avg, colorlist=self.colorlist)
        
        # quick plotting ground truth
        multiple_variants_plot(title=f"{self.thetitle} (ground truth)", xlab=self.thexlabel, ylab=self.theylabel, variants_name=self.variants_name_groundtruth, 
                                x_list=self.pdb_groundtruth["x_list"], y_list=self.pdb_groundtruth["y_list"], ax=ax3,
                                x_avg_list=self.pdb_groundtruth["x_avg_list"], y_avg_list=self.pdb_groundtruth["y_avg_list"], 
                                do_moving_avg=False, alpha=0.15, colorlist=self.colorlist_groundtruth, showlabel=False)
        multiple_variants_plot(title=f"{self.thetitle} (ground truth)", xlab=self.thexlabel, ylab=self.theylabel, variants_name=self.variants_name_groundtruth, 
                                x_list=self.pdb_7a91["x_list"], y_list=self.pdb_7a91["y_list"], ax=ax3,
                                x_avg_list=self.pdb_groundtruth["x_avg_list"], y_avg_list=self.pdb_groundtruth["y_avg_list"], 
                                do_moving_avg=self.do_moving_avg, colorlist=self.colorlist_groundtruth)

    def relationship_plot(self, min_value, max_value):
        fig = plt.figure(figsize=[14,8])
        ax1 = fig.add_subplot(111)

        ax1.set_title(f"Relationship ({theylabel}) between 7a91 and 6m0j ")
        ax1.set_xlabel("7a91")
        ax1.set_ylabel("6m0j")

        ax1.ticklabel_format(axis="y")
        for index,eachname in enumerate(variants_name):
            ax1.plot(self.dfs[0].T[eachname], self.dfs[1].T[eachname], label=eachname, color = self.colorlist[index])
        ax1.legend()
        ax1.axis(xmin=min_value,xmax=max_value, ymin=min_value,ymax=max_value)
```
